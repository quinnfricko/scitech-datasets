```python
#Imports Pandas and Matplotlib as well as CSV
# WINTER OLYMPICS STATS

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/workspaces/scitech-datasets/WinterSD.csv')
```

```python
#CHART FOR US MEDALS SORTED BY TYPE

fig, ax = plt.subplots()


# Filters ony US Medals
usa_medals = df[df['Country'] == 'USA']

# Sorts each medal type
gold_count = len(usa_medals[usa_medals['Medal'] == 'Gold'])
silver_count = len(usa_medals[usa_medals['Medal'] == 'Silver'])
bronze_count = len(usa_medals[usa_medals['Medal'] == 'Bronze'])

# Sets parameters for chart
medals = ['Gold', 'Silver', 'Bronze']
total_counts = [gold_count, silver_count, bronze_count]
bar_colors = ['gold', 'silver', 'tan']

ax.bar(medals, total_counts, color=bar_colors)

ax.set_ylabel('Count')
ax.set_xlabel('Medal Type')

ax.set_title('All Time Medals Earned by United States in the Winter Olympics')

#Show Chart
plt.show()



#Ref : https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_colors.html
```

```python
#CHART FOR US MEDALS BY YEARr

fig, ax = plt.subplots()

# Filter only USA medals
usa_medals = df[df['Country'] == 'USA']

# Count medals per year
medals_by_year = usa_medals.groupby('Year').size()

# Plot  the line chart
ax.plot(medals_by_year.index, medals_by_year.values,
        marker='o', linestyle='--')


#label labels
ax.set_xlabel('Year')
ax.set_ylabel('Total Medals')
ax.set_title('United States Winter Olympic Medals by Year')

plt.show()


# ref : https://matplotlib.org/stable/gallery/lines_bars_and_markers/line_demo_dash_control.html
```
